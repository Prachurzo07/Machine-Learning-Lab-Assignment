```python
import numpy as np
import pandas as pd
```

```python
# 1. Create two linear arrays a and b with 20 random elements (1 to 100)

a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

print("Array a:\n", a)
print("Array b:\n", b)
```

```python
# 2. Convert linear arrays to 5x4 matrices

a = a.reshape(5, 4)  # 5 rows and 4 columns
b = b.reshape(5, 4)

print("\nMatrix a:\n", a)
print("\nMatrix b:\n", b)
```

```python
# 3. Perform scalar multiplication
value = 10
a = a * value
b = b * value

print("\nMatrix a after scalar multiplication:\n", a)
print("\nMatrix b after scalar multiplication:\n", b)
```

```python
# 4. Modify matrix b and perform matrix multiplication

b_modified = b.reshape(4, 5)

c = np.matmul(a, b_modified)

print("\nMatrix b modified (4x5):\n", b_modified)
print("\nMatrix c (a x b_modified):\n", c)
```

```python
# 5. Print only column 2 and 3 of matrix c (index 1 and 2)

print("\nColumns 2 and 3 of matrix c:\n", c[:, 1:3])

# Print whole matrix using slicing
print("\nWhole matrix c using slicing:\n", c[:, :])
```

```python
# 6. Find largest element and its index in c

largest = c.max()
row, col = np.unravel_index(np.argmax(c), c.shape)  # First it finds the index in 1D then it converts that 1D index into the 2D formate index

position = (int(row), int(col))

print("Largest element:", largest)
print("Index (row, column):", position)
```

```python
# 7. Convert matrix c into one-dimensional array (without changing original)
c_flat = c.flatten()

print("\nMatrix c converted to 1D array:\n", c_flat)
```

```python
import matplotlib.pyplot as plt


temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]

study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks_line = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_bar = [89, 90, 45, 78, 99]

plt.figure(figsize=(14,10))



plt.subplot(2,2,1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")



plt.subplot(2,2,2)
plt.plot(study_hours, marks_line)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")


plt.subplot(2,2,3)
plt.barh(subjects, marks_bar)
plt.xlabel("Marks")
plt.title("Subject Scores (Horizontal)")


plt.subplot(2,2,4)
colors = ['blue', 'green', 'orange', 'purple', 'red']
plt.bar(subjects, marks_bar, color=colors)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Subject Scores (Vertical)")



plt.tight_layout()
plt.show()
```
